```python
%%time
from numba import jit

@jit(nopython=True)
def createOrderingCalendar(g_lNumProdStoreRows, g_lXTabColEndPlanHorizExt,
                           g_lXTabColToday, g_lStoreRowForProdStoreRow, 
                           g_lFestivo, g_lDateForXTabCol_dayofweek, g_lPOSendDayForStoreRow,
                           g_lLeadTimeForStoreRow, g_lNumWorkingDaysForStoreRow,
                           g_lCarrierDayOff,lTgtProdStoreRow = 0):
    # --------------------------------
    # Funcion: CreateOrderingCalendar
    # --------------------------------
    # Esta función crea el calendario de días en los que se pueden hacer pedidos por cada producto-tienda
    # Es una función que puede utilizarse a demanda dentro de la herramienta.

#     global g_lNumProdStoreRows, g_lRecXTabColForProdStoreRowXTabCol, g_lXTabColEndPlanHorizExt, g_lXTabColToday, g_lStoreRowForProdStoreRow, g_lFestivo, \
#         g_lDateForXTabCol_dayofweek, g_lPOSendDayForStoreRow, g_lLeadTimeForStoreRow, g_lNumWorkingDaysForStoreRow, g_lCarrierDayOff
    
    """
        When using numba all the parameters should not 
        contain strings, dictionaries, classes, or in general, python objects.
        For all the functions here, only numpy object were used and some 
        integers too.

    """    
    g_lLeadTimeForStoreRow = g_lLeadTimeForStoreRow.astype(np.int32)
    g_lStoreRowForProdStoreRow = g_lStoreRowForProdStoreRow.astype(np.int32)
    
    if lTgtProdStoreRow == 0:
#         for i in range(g_lNumProdStoreRows):
#             g_lRecXTabColForProdStoreRowXTabCol.append([0] * (g_lXTabColEndPlanHorizExt + 1))
            
        g_lRecXTabColForProdStoreRowXTabCol = np.zeros((g_lNumProdStoreRows, g_lXTabColEndPlanHorizExt+1))
    else:
        for lXTabCol in np.arange(g_lXTabColToday , g_lXTabColEndPlanHorizExt + 1):
            g_lRecXTabColForProdStoreRowXTabCol[lTgtProdStoreRow-1][lXTabCol] = 0

    # lProdStoreRow = 2
    for lProdStoreRow in np.arange(1, g_lNumProdStoreRows + 1):
        if lTgtProdStoreRow == 0 or lProdStoreRow == lTgtProdStoreRow:
            # print(lProdStoreRow)
            lStoreRow = g_lStoreRowForProdStoreRow[lProdStoreRow - 1]
            # print(lStoreRow)
            # lXTabCol = g_lXTabColToday
            for lXTabCol in np.arange(g_lXTabColToday, g_lXTabColEndPlanHorizExt + 1):
                # print(lXTabCol)
                # Check DC is working this day
                if g_lFestivo[lXTabCol] == 0:
                    # print(g_lFestivo[lXTabCol])
#                     lDay = datetime.date(g_lDateForXTabCol[lXTabCol]).weekday()  # en python, 0:Lunes, 1:Martes, ... 6:Domingo
                    lDay = g_lDateForXTabCol_dayofweek[lXTabCol]  # en python, 0:Lunes, 1:Martes, ... 6:Domingo
                    # print(lDay)
                    if g_lPOSendDayForStoreRow[lStoreRow - 1][lDay] == 1:  # este día se permite hacer pedidos

                        # Add lead time
                        lXTabColRec = lXTabCol + g_lLeadTimeForStoreRow[lStoreRow - 1]  # se calcula que dia llegara el pedido a tienda

                        # If store only open 5 days per week then reschedule Saturday or Sunday deliveries for Monday
                        if g_lNumWorkingDaysForStoreRow[lStoreRow - 1] == 5:
#                             lDay = datetime.date(g_lDateForXTabCol[lXTabColRec]).weekday()
                            lDay = g_lDateForXTabCol_dayofweek[lXTabColRec]
                            if lDay == 5: lXTabColRec = lXTabColRec + 2  # Sabado
                            if lDay == 6: lXTabColRec = lXTabColRec + 1  # Domingo

                        # If store only open 6 days per week then reschedule Sunday deliveries for Monday
                        if g_lNumWorkingDaysForStoreRow[lStoreRow - 1] == 6:
#                             lDay = datetime.date(g_lDateForXTabCol[lXTabColRec]).weekday()
                            lDay = g_lDateForXTabCol_dayofweek[lXTabColRec]
                            if lDay == 6: lXTabColRec = lXTabColRec + 1  # Domingo

                        # If not a working day for carrier then keep adding days until it is a receivable day
                        if g_lCarrierDayOff[lXTabColRec] == 1:
                            while g_lCarrierDayOff[lXTabColRec] != 0:
                                lXTabColRec = lXTabColRec + 1

                        # Store receipt day for PO day
                        g_lRecXTabColForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] = lXTabColRec
                        
    return g_lRecXTabColForProdStoreRowXTabCol

# g_lRecXTabColForProdStoreRowXTabCol = createOrderingCalendar(g_lNumProdStoreRows, g_lXTabColEndPlanHorizExt,
#                            g_lXTabColToday, g_lStoreRowForProdStoreRow, 
#                            g_lFestivo, g_lDateForXTabCol, g_lPOSendDayForStoreRow,
#                            g_lLeadTimeForStoreRow, g_lNumWorkingDaysForStoreRow,
#                            g_lCarrierDayOff,lTgtProdStoreRow = 0)


# g_lFestivo = g_lFestivo.astype(np.int32)
# g_lCarrierDayOff = g_lCarrierDayOff.astype(np.int32)
g_lRecXTabColForProdStoreRowXTabCol = createOrderingCalendar(g_lNumProdStoreRows, g_lXTabColEndPlanHorizExt,
                           g_lXTabColToday, g_lStoreRowForProdStoreRow, 
                           g_lFestivo, g_lDateForXTabCol_dayofweek, g_lPOSendDayForStoreRow,
                           g_lLeadTimeForStoreRow, g_lNumWorkingDaysForStoreRow,
                           g_lCarrierDayOff,lTgtProdStoreRow = 0)
```

# Funcion: CALCULO DEL EOQ **

- Sometimes can happen that a whole function can not be rewritten in a numba way.
- In that case, we can rewrite a portion of that function and call it inside the parent function.


- Original: 1.2 s

- Optimized: 1.3 s

- 0.92 times faster than original

```python
%%time
from numba import jit

@jit(nopython=True)
def calculateEOQ_temp(lProdStoreRow, g_lXTabColToday, g_lXTabColEndPlanHorizExt, g_dPlanDemForProdStoreRowXTabCol):
    dAnnualDemand = 0
    for lXTabCol in np.arange(g_lXTabColToday, g_lXTabColEndPlanHorizExt + 1):
        dAnnualDemand = dAnnualDemand + g_dPlanDemForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
    return dAnnualDemand

def calculateEOQ(g_lNumProdStoreRows, g_lXTabColToday, g_lXTabColEndPlanHorizExt,
                g_dPlanDemForProdStoreRowXTabCol, g_lProdRowForProdStoreRow, 
                g_dictConvFactorForProdUOM1UOM2, g_productDims, g_dictUOMCDForProd, 
                g_dictMinUOMForProd, g_lUnitCostDefaultForProdRow, g_lMinLotSizeForProdStoreRow):
    # -------------------------
    # Funcion: CALCULO DEL EOQ
    # -------------------------
#     global g_lLotSizeForProdStoreRow, g_lEOQForProdStoreRow, g_lNumProdStoreRows, g_lXTabColToday, g_lXTabColEndPlanHorizExt, g_dPlanDemForProdStoreRowXTabCol,\
#         g_lProdRowForProdStoreRow, g_dictConvFactorForProdUOM1UOM2, g_productDims, g_dictUOMCDForProd, g_dictMinUOMForProd, g_lUnitCostDefaultForProdRow,\
#         g_lMinLotSizeForProdStoreRow

#     for i in range(g_lNumProdStoreRows):
#         g_lLotSizeForProdStoreRow.append(0)
#         g_lEOQForProdStoreRow.append(0)

    g_lLotSizeForProdStoreRow = np.zeros(g_lNumProdStoreRows)
    g_lEOQForProdStoreRow = np.zeros(g_lNumProdStoreRows)
    
    # lProdStoreRow = 1
    for lProdStoreRow in np.arange(1, g_lNumProdStoreRows + 1):
        dEOQ = 0
        dAnnualDemand = 0
        # demanda anual
        # lProdStoreRow = 2
#         for lXTabCol in np.arange(g_lXTabColToday, g_lXTabColEndPlanHorizExt + 1):
#             dAnnualDemand = dAnnualDemand + g_dPlanDemForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
        """
            Here a part of the function was embeded. The most expensive. 

        """   
        dAnnualDemand = calculateEOQ_temp(lProdStoreRow, g_lXTabColToday, g_lXTabColEndPlanHorizExt,g_dPlanDemForProdStoreRowXTabCol)
        dAnnualDemand = 364 * (dAnnualDemand / (g_lXTabColEndPlanHorizExt - g_lXTabColToday + 1))

        lProdRow = g_lProdRowForProdStoreRow[lProdStoreRow - 1]

        # obteniendo el costo unitario en unidades minimas para el pdv
        # sProd = g_productDims[lProdRow-1][0]
        # uCD = g_dictUOMCDForProd[sProd]
        # uMin = g_dictMinUOMForProd[sProd]
        # skey = sProd + uCD + uMin
        # factor = g_dictConvFactorForProdUOM1UOM2[skey]

        g_productDims_index = int(g_productDims[lProdRow - 1][0])
        factor = g_dictConvFactorForProdUOM1UOM2[
            str(g_productDims_index) + g_dictUOMCDForProd[g_productDims_index] + g_dictMinUOMForProd[g_productDims_index]]
#         print(factor)
        lUnitCost = g_lUnitCostDefaultForProdRow[lProdRow - 1] / factor

        if lUnitCost > 0:
            dEOQ = math.sqrt((2 * cte.lote_costo_de_pedir * dAnnualDemand) / (lUnitCost * cte.lote_nual_holding_cost))

        g_lEOQForProdStoreRow[lProdStoreRow - 1] = int(dEOQ)  # Se trunca el EOQ

        # Ajustando a los parámetros de lote:
        dMaxLotSizeUnits = 0
        dMinLotSizeUnits = 0

        # Find max lot size
        dMaxLotSizeUnits = cte.lote_cantidad_maximaa_pedir_semanas * (dAnnualDemand / 52)
        # Find min lote size
        dMinLotSizeUnits = cte.lote_cantidad_minima_pedir_semanas * (dAnnualDemand / 52)

        if lUnitCost > 0:
            if dMaxLotSizeUnits > (cte.lote_valor_maximo_envio_x_linea / lUnitCost):
                dMaxLotSizeUnits = (cte.lote_valor_maximo_envio_x_linea / lUnitCost)

        # Apply caps for months and value, and apply floor of min lot size
        # 1) Lote Size = EOQ
        g_lLotSizeForProdStoreRow_tmp = g_lLotSizeForProdStoreRow[lProdStoreRow - 1]
        
        g_lLotSizeForProdStoreRow_tmp = g_lEOQForProdStoreRow[lProdStoreRow - 1]
        # 2) El tamaño del Lote no puede ser mayor que el Máximo definido
        if g_lLotSizeForProdStoreRow_tmp > dMaxLotSizeUnits:
            g_lLotSizeForProdStoreRow_tmp = int(dMaxLotSizeUnits)
        # 3) El tamaño del Lote no puede ser menor que el Mínimo definido
        if g_lLotSizeForProdStoreRow_tmp < dMinLotSizeUnits:
            g_lLotSizeForProdStoreRow_tmp = int(dMinLotSizeUnits)
        # 4) El tamaño del Lote no puede ser menor que el Lote mínimo del Producto-Tienda
        g_lMinLotSizeForProdStoreRow_tmp = g_lMinLotSizeForProdStoreRow[lProdStoreRow - 1]
        if g_lLotSizeForProdStoreRow_tmp < g_lMinLotSizeForProdStoreRow_tmp:
            g_lLotSizeForProdStoreRow_tmp = g_lMinLotSizeForProdStoreRow_tmp
        # El tamaño del lote se redondea a un múltiplo del Lote minimo que sea menor a él mismo
        if g_lMinLotSizeForProdStoreRow_tmp > 0:
            g_lLotSizeForProdStoreRow_tmp = g_lMinLotSizeForProdStoreRow_tmp * int(
                0.999999 + g_lLotSizeForProdStoreRow_tmp / g_lMinLotSizeForProdStoreRow_tmp)
            
        g_lLotSizeForProdStoreRow[lProdStoreRow - 1] = g_lLotSizeForProdStoreRow_tmp
        
    return g_lLotSizeForProdStoreRow, g_lEOQForProdStoreRow     

g_lLotSizeForProdStoreRow,g_lEOQForProdStoreRow  = calculateEOQ(g_lNumProdStoreRows, g_lXTabColToday, g_lXTabColEndPlanHorizExt,
                                                    g_dPlanDemForProdStoreRowXTabCol, g_lProdRowForProdStoreRow, 
                                                    g_dictConvFactorForProdUOM1UOM2, g_productDims, g_dictUOMCDForProd, 
                                                    g_dictMinUOMForProd, g_lUnitCostDefaultForProdRow, g_lMinLotSizeForProdStoreRow)
```

```python
%%time
from numba import jit

@jit(nopython=True)
def calculateTargetStock(g_lNumProdStoreRows,
                         g_lXTabColEndPlanHorizExt,
                         g_lDaysCoverForProdStoreRow,
                         g_lXTabColToday,
                         g_dPlanDemForProdStoreRowXTabCol):
# -------------------------------
# funcion: CALCULATE TARGET STOCK
# --------------------------------
# Calcula sl stock objetivo por cada producto-tienda

#     global g_lTgtStockForProdStoreRowXTabCol, g_lNumProdStoreRows, g_lXTabColEndPlanHorizExt, g_lDaysCoverForProdStoreRow, g_lXTabColToday, \
#         g_dPlanDemForProdStoreRowXTabCol

    # for i in range(g_lNumProdStoreRows):
    #     g_lTgtStockForProdStoreRowXTabCol.append([0] * (g_lXTabColEndPlanHorizExt + 1))

    g_lTgtStockForProdStoreRowXTabCol = np.zeros((g_lNumProdStoreRows, g_lXTabColEndPlanHorizExt+1))

    """
      np.arange() can work like range() and it is compatible with numba.

    """ 
    # lProdStoreRow = 1
#     for lProdStoreRow in range(1, g_lNumProdStoreRows + 1):
    for lProdStoreRow in np.arange(1, g_lNumProdStoreRows + 1):
        # print(lProdStoreRow)
        # lStoreRow = g_lStoreRowForProdStoreRow[lProdStoreRow-1]
        # lProdRow = g_lProdRowForProdStoreRow[lProdStoreRow-1]

        lDOS = g_lDaysCoverForProdStoreRow[lProdStoreRow - 1]  # cobertura
        # lXTabCol = g_lXTabColToday
#         for lXTabCol in range(g_lXTabColToday, (g_lXTabColEndPlanHorizExt + 1)):
        for lXTabCol in np.arange(g_lXTabColToday, (g_lXTabColEndPlanHorizExt + 1)):
            dDemandDOS = 0
#             for lXTabColAux in range(lXTabCol, lXTabCol + int(lDOS) - 1 + 1):
            for lXTabColAux in np.arange(lXTabCol, lXTabCol + int(lDOS) - 1 + 1):
                # print(lXTabColAux)
                if lXTabColAux > g_lXTabColEndPlanHorizExt:
                    # Se repite la última semana
                    lDay = lXTabColAux - g_lXTabColEndPlanHorizExt
                    lSem = int(lDay / 7)
                    dDemandDOS = dDemandDOS + g_dPlanDemForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabColAux - 7 * (1 + lSem)]
                    # dDemandDOS = 0
                else:
                    dDemandDOS = dDemandDOS + g_dPlanDemForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabColAux]

            g_lTgtStockForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] = int(0.5 + dDemandDOS)
            
    return g_lTgtStockForProdStoreRowXTabCol

g_lTgtStockForProdStoreRowXTabCol = calculateTargetStock( 
                                             g_lNumProdStoreRows,
                                             g_lXTabColEndPlanHorizExt,
                                             g_lDaysCoverForProdStoreRow,
                                             g_lXTabColToday,
                                             g_dPlanDemForProdStoreRowXTabCol)
```

```python
%%time
from numba import jit

@jit(nopython=True)
def RunDRP(g_lNumProdStoreRows, g_lXTabColEndPlanHoriz,
           g_lXTabColToday, g_lStoreRowForProdStoreRow, g_lProdRowForProdStoreRow,
           g_lStartingInTransitForProdStoreRow, g_lStartingInvForProdStoreRow, 
           g_lSchedRecForProdStoreRowXTabCol,g_lRecXTabColForProdStoreRowXTabCol,
           g_lXTabColEndPlanHorizExt, g_dPlanDemForProdStoreRowXTabCol,
           g_lTgtStockForProdStoreRowXTabCol, g_lLotSizeForProdStoreRow,
           g_lMinLotSizeForProdStoreRow, lTgtProdStoreRow=0):
    """
    FUNCION Sub RunDRP(lTgtProdStoreRow As Long)
    """

#     global g_lPlanOrdForProdStoreRowXTabCol, g_lPlanRecForProdStoreRowXTabCol, g_lPlanSalesForProdStoreRowXTabCol, g_lPlanRefusForProdStoreRowXTabCol,\
#         g_lPlanInvForProdStoreRowXTabCol, g_lPlanInTransitForProdStoreRowXTabCol, g_lPlanDispForProdStoreRowXTabCol, g_lNumProdStoreRows,\
#         g_lXTabColEndPlanHoriz, g_lXTabColToday, g_lStoreRowForProdStoreRow, g_lProdRowForProdStoreRow, g_lStartingInTransitForProdStoreRow,\
#         g_lStartingInvForProdStoreRow, g_lSchedRecForProdStoreRowXTabCol,g_lRecXTabColForProdStoreRowXTabCol, g_lXTabColEndPlanHorizExt, \
#         g_dPlanDemForProdStoreRowXTabCol, g_lTgtStockForProdStoreRowXTabCol, g_lLotSizeForProdStoreRow, g_lMinLotSizeForProdStoreRow

    if lTgtProdStoreRow == 0:
        g_lPlanOrdForProdStoreRowXTabCol = np.zeros((g_lNumProdStoreRows, g_lXTabColEndPlanHoriz+1))
        g_lPlanRecForProdStoreRowXTabCol = np.zeros((g_lNumProdStoreRows, g_lXTabColEndPlanHoriz+1))
        g_lPlanSalesForProdStoreRowXTabCol = np.zeros((g_lNumProdStoreRows, g_lXTabColEndPlanHoriz+1))
        g_lPlanRefusForProdStoreRowXTabCol = np.zeros((g_lNumProdStoreRows, g_lXTabColEndPlanHoriz+1))
        g_lPlanInvForProdStoreRowXTabCol = np.zeros((g_lNumProdStoreRows, g_lXTabColEndPlanHoriz+1))
        g_lPlanInTransitForProdStoreRowXTabCol = np.zeros((g_lNumProdStoreRows, g_lXTabColEndPlanHoriz+1))
        g_lPlanDispForProdStoreRowXTabCol = np.zeros((g_lNumProdStoreRows, g_lXTabColEndPlanHoriz+1))
        
#         for i in range(g_lNumProdStoreRows):
#             g_lPlanOrdForProdStoreRowXTabCol.append([0] * (g_lXTabColEndPlanHoriz + 1))
#             g_lPlanRecForProdStoreRowXTabCol.append([0] * (g_lXTabColEndPlanHoriz + 1))
#             g_lPlanSalesForProdStoreRowXTabCol.append([0] * (g_lXTabColEndPlanHoriz + 1))
#             g_lPlanRefusForProdStoreRowXTabCol.append([0] * (g_lXTabColEndPlanHoriz + 1))
#             g_lPlanInvForProdStoreRowXTabCol.append([0] * (g_lXTabColEndPlanHoriz + 1))
#             g_lPlanInTransitForProdStoreRowXTabCol.append([0] * (g_lXTabColEndPlanHoriz + 1))
#             g_lPlanDispForProdStoreRowXTabCol.append([0] * (g_lXTabColEndPlanHoriz + 1))
    else:
        for lXTabCol in np.arange(g_lXTabColToday , g_lXTabColEndPlanHoriz + 1):
            g_lPlanOrdForProdStoreRowXTabCol[lTgtProdStoreRow-1][lXTabCol] = 0
            g_lPlanRecForProdStoreRowXTabCol[lTgtProdStoreRow-1][lXTabCol] = 0
            g_lPlanSalesForProdStoreRowXTabCol[lTgtProdStoreRow-1][lXTabCol] = 0
            g_lPlanRefusForProdStoreRowXTabCol[lTgtProdStoreRow-1][lXTabCol] = 0
            g_lPlanInvForProdStoreRowXTabCol[lTgtProdStoreRow-1][lXTabCol] = 0
            g_lPlanInTransitForProdStoreRowXTabCol[lTgtProdStoreRow-1][lXTabCol] = 0
            g_lPlanDispForProdStoreRowXTabCol[lTgtProdStoreRow-1][lXTabCol] = 0

    # lProdStoreRow = 1
    for lProdStoreRow in np.arange(1, g_lNumProdStoreRows + 1):
        if lTgtProdStoreRow == 0 or lProdStoreRow == lTgtProdStoreRow:
            lStoreRow = g_lStoreRowForProdStoreRow[lProdStoreRow - 1]
            lProdRow = g_lProdRowForProdStoreRow[lProdStoreRow - 1]

            if lStoreRow > 0:
                lCumInTransit = g_lStartingInTransitForProdStoreRow[lProdStoreRow - 1]
                g_lPlanInvForProdStoreRowXTabCol[lProdStoreRow - 1][g_lXTabColToday - 1] = g_lStartingInvForProdStoreRow[lProdStoreRow - 1] + \
                                                                                           g_lSchedRecForProdStoreRowXTabCol[lProdStoreRow - 1][g_lXTabColToday - 1]
                g_lPlanInTransitForProdStoreRowXTabCol[lProdStoreRow - 1][g_lXTabColToday - 1] = lCumInTransit

                # lXTabCol = g_lXTabColToday+4
                for lXTabCol in np.arange(g_lXTabColToday, g_lXTabColEndPlanHoriz + 1):
                    # if lXTabCol == 76:
                    #     print("ErroR")
                    # print(lXTabCol)
                    # aqui van las ordenes manuales que no conisederé para la demo
                    # y luego las ordenes automaticas

                    # 'Check we can place orders on this day
                    lRecXTabColTHISOrder = g_lRecXTabColForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
                    if lRecXTabColTHISOrder > 0:
                        # Find receipt day for NEXT order
                        lXTabColNEXTOrder = 0
                        for lXTabColNEXTOrder in np.arange(lXTabCol + 1, (g_lXTabColEndPlanHorizExt + 1)):
                            if g_lRecXTabColForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabColNEXTOrder] > 0:
                                break

                        if lXTabColNEXTOrder <= g_lXTabColEndPlanHorizExt:
                            lRecXTabColNEXTOrder = g_lRecXTabColForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabColNEXTOrder]
                        else:
                            lRecXTabColNEXTOrder = 0

                        # Cumulate Plan Dem  int(g_productDims[int(lProdRow) - 1][0])
                        lCumPlanDem = 0
                        for lTgtXTabCol in np.arange(lXTabCol, min((g_lXTabColEndPlanHorizExt + 1), lRecXTabColNEXTOrder - 1)):
                            lCumPlanDem = lCumPlanDem + g_dPlanDemForProdStoreRowXTabCol[lProdStoreRow - 1][int(lTgtXTabCol)]

                        # Recommend Order planeado
                        g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] = max(0, (
                                lCumPlanDem - lCumInTransit - g_lPlanInvForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol - 1] +
                                g_lTgtStockForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]))

                        # apply lote size
                        if g_lLotSizeForProdStoreRow[lProdStoreRow - 1] > 1:
                            if g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] > 0:
                                # Enforce lot size
                                if g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] < g_lLotSizeForProdStoreRow[lProdStoreRow - 1]:
                                    g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] = g_lLotSizeForProdStoreRow[lProdStoreRow - 1]
                                # Round up to multiple of min lot size
                                if g_lMinLotSizeForProdStoreRow[lProdStoreRow - 1] > 0:
                                    g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] = g_lMinLotSizeForProdStoreRow[lProdStoreRow - 1] * int(
                                        0.999999 + g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] / g_lMinLotSizeForProdStoreRow[lProdStoreRow - 1])

                        # Plan Receipt
                        if lRecXTabColTHISOrder <= g_lXTabColEndPlanHoriz:
                            g_lPlanRecForProdStoreRowXTabCol[lProdStoreRow - 1][int(lRecXTabColTHISOrder)] = g_lPlanRecForProdStoreRowXTabCol[lProdStoreRow - 1][int(lRecXTabColTHISOrder)] + g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][int(lXTabCol)]

                        # Update cumulative on order
                        lCumInTransit = lCumInTransit + g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]

                    # Update On Order, dispatches, inventory and refusals
                    lCumInTransit = lCumInTransit - g_lSchedRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] - g_lPlanRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
                    g_lPlanInTransitForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] = lCumInTransit

                    g_lPlanDispForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] = min(g_lPlanInvForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol - 1] + g_lSchedRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] + g_lPlanRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol], g_dPlanDemForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol])
                    g_lPlanInvForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] = g_lPlanInvForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol - 1] + g_lSchedRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] + g_lPlanRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] - int(g_lPlanDispForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol])
                    g_lPlanRefusForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] = int(g_dPlanDemForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]) - int(g_lPlanDispForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol])
    
    return g_lPlanOrdForProdStoreRowXTabCol, g_lPlanRecForProdStoreRowXTabCol, g_lPlanSalesForProdStoreRowXTabCol, \
            g_lPlanRefusForProdStoreRowXTabCol, g_lPlanInvForProdStoreRowXTabCol, g_lPlanInTransitForProdStoreRowXTabCol, \
            g_lPlanDispForProdStoreRowXTabCol              
        
g_lPlanOrdForProdStoreRowXTabCol, g_lPlanRecForProdStoreRowXTabCol, g_lPlanSalesForProdStoreRowXTabCol, \
            g_lPlanRefusForProdStoreRowXTabCol, g_lPlanInvForProdStoreRowXTabCol, g_lPlanInTransitForProdStoreRowXTabCol, \
            g_lPlanDispForProdStoreRowXTabCol = RunDRP(g_lNumProdStoreRows, g_lXTabColEndPlanHoriz,
                                                   g_lXTabColToday, g_lStoreRowForProdStoreRow, g_lProdRowForProdStoreRow,
                                                   g_lStartingInTransitForProdStoreRow, g_lStartingInvForProdStoreRow, 
                                                   g_lSchedRecForProdStoreRowXTabCol,g_lRecXTabColForProdStoreRowXTabCol,
                                                   g_lXTabColEndPlanHorizExt, g_dPlanDemForProdStoreRowXTabCol,
                                                   g_lTgtStockForProdStoreRowXTabCol, g_lLotSizeForProdStoreRow,
                                                   g_lMinLotSizeForProdStoreRow, lTgtProdStoreRow=0)
```

```python
%%time
from numba import jit

@jit(nopython=True)
def planeación_detallada_temp(g_lXTabColToday, g_lXTabColEndPlanHoriz,
                             lTotSchedRecUU, g_lSchedRecForProdStoreRowXTabCol,
                             lTotPlanRecUU, g_lPlanRecForProdStoreRowXTabCol, 
                             lProdStoreRow, lTotDispUU, g_lPlanOrdForProdStoreRowXTabCol,
                             dFactor, lTotDispPesos, dUnitCost, lTotRefusUU, lTotInvUU,
                             g_lPlanInvForProdStoreRowXTabCol, lTotDispLineas, dTotInvPesos,
                             dTotInTransitPesos):
    
    for lXTabCol in np.arange(g_lXTabColToday, g_lXTabColEndPlanHoriz + 1):
        # TOTALES
        # -------

        # Entradas confirmadas
        lTotSchedRecUU[lXTabCol] = lTotSchedRecUU[lXTabCol] + g_lSchedRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
        # Entradas PlaneadaS
        lTotPlanRecUU[lXTabCol] = lTotPlanRecUU[lXTabCol] + g_lPlanRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
        # Pedidos Planeados O Despachos Planeados en Unidades Cedi
        if dFactor > 0:
            lTotDispUU[lXTabCol] = lTotDispUU[lXTabCol] + g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] / dFactor

        # Total Distribucion en $MM
        lTotDispPesos[lXTabCol] = lTotDispPesos[lXTabCol] + g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] * (dUnitCost / 1000000)
        # Venta Perdida
        lTotRefusUU[lXTabCol] = lTotRefusUU[lXTabCol] + g_lPlanRefusForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
        # Stock On Hand
        lTotInvUU[lXTabCol] = lTotInvUU[lXTabCol] + g_lPlanInvForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
        # Total Lineas
        if g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] > 0:
            lTotDispLineas[lXTabCol] = lTotDispLineas[lXTabCol] + 1

        # Total de Inventarios en manos en $MM
        dTotInvPesos[lXTabCol] = dTotInvPesos[lXTabCol] + g_lPlanInvForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] * dUnitCost
        # Total de Inventarios en transito en $MM
        dTotInTransitPesos[lXTabCol] = dTotInTransitPesos[lXTabCol] + g_lPlanInTransitForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] * dUnitCost
        
    return lTotSchedRecUU, lTotPlanRecUU, lTotDispUU, lTotDispPesos, lTotRefusUU, lTotInvUU, lTotDispLineas, dTotInvPesos, dTotInTransitPesos

def RealityCheck(g_prodStoreRowSelected, g_lDateForXTabCol_dayofweek, 
                 g_lDateForXTabCol_day, g_lDateForXTabCol_month,
                 g_lXTabColToday, g_lXTabColEndPlanHoriz, g_lNumProdStoreRows, 
                 g_productDims, g_lSchedRecForProdStoreRowXTabCol, 
                 g_lPlanRecForProdStoreRowXTabCol, g_lPlanOrdForProdStoreRowXTabCol,
                 g_lPlanRefusForProdStoreRowXTabCol, g_lPlanInvForProdStoreRowXTabCol,
                 g_lPlanInTransitForProdStoreRowXTabCol):
    """
    FUNCION RealityCheck()
    para independizar
    Esta función genera la hoja "Planeación Detallada" de la herramienta en version Excel
    Nota: no confundir el nombre con el Reality Check verdadero, allan colocó este nombre en la herramienta inicial y se mantuvo hasta el final
    """
#     global g_prodStoreRowSelected, g_lDateForXTabCol_dayofweek, g_lDateForXTabCol_day, g_lDateForXTabCol_month, g_lXTabColToday, g_lXTabColEndPlanHoriz, g_lNumProdStoreRows

#     global df_vDRPTitleBlocks, df_vDRPBlocks, df_vColTotals,df_vColHeaders # para fines didacticos

    lMaxDRPTitleBlocks = 5
    lNumRowsPerDRPTitleBlock = 10

    # Set up output arrays
    # --------------------
    vDRPTitleBlocks = []
    vDRPBlocks = []
    booOverride = []

    for i in range(lMaxDRPTitleBlocks * (lNumRowsPerDRPTitleBlock + 1) - 1):
        vDRPTitleBlocks.append([''] * (3))
        vDRPBlocks.append([0] * (g_lXTabColEndPlanHoriz + 1))
        booOverride.append([0] * (g_lXTabColEndPlanHoriz + 1))

#     vDRPTitleBlocks = np.empty([lMaxDRPTitleBlocks * (lNumRowsPerDRPTitleBlock + 1) - 1, 3], dtype="S40")
#     vDRPBlocks = np.zeros((lMaxDRPTitleBlocks * (lNumRowsPerDRPTitleBlock + 1) - 1, g_lXTabColEndPlanHoriz + 1))
#     booOverride = np.zeros((lMaxDRPTitleBlocks * (lNumRowsPerDRPTitleBlock + 1) - 1, g_lXTabColEndPlanHoriz + 1))
    # Set up arrays for totals
    # ------------------------
#     lTotSchedRecUU = []
#     dTotInvPesos = []
#     lTotDispUU = []
#     lTotDispPesos = []
#     lTotInvUU = []
#     lTotPlanRecUU = []
#     dTotInTransitPesos = []
#     lTotDispLineas = []
#     lTotRefusUU = []

    lTotSchedRecUU = np.zeros(g_lXTabColEndPlanHoriz + 1)
    dTotInvPesos = np.zeros(g_lXTabColEndPlanHoriz + 1)
    lTotDispUU = np.zeros(g_lXTabColEndPlanHoriz + 1)
    lTotDispPesos = np.zeros(g_lXTabColEndPlanHoriz + 1)
    lTotInvUU = np.zeros(g_lXTabColEndPlanHoriz + 1)
    lTotPlanRecUU = np.zeros(g_lXTabColEndPlanHoriz + 1)
    dTotInTransitPesos = np.zeros(g_lXTabColEndPlanHoriz + 1)
    lTotDispLineas = np.zeros(g_lXTabColEndPlanHoriz + 1)
    lTotRefusUU = np.zeros(g_lXTabColEndPlanHoriz + 1)
    
#     for i in range(g_lXTabColEndPlanHoriz + 1):
#         lTotSchedRecUU.append(0)
#         dTotInvPesos.append(0)
#         lTotDispUU.append(0)
#         lTotDispPesos.append(0)
#         lTotInvUU.append(0)
#         lTotPlanRecUU.append(0)
#         dTotInTransitPesos.append(0)
#         lTotDispLineas.append(0)
#         lTotRefusUU.append(0)

#     vColTotals = np.zeros((4, g_lXTabColEndPlanHoriz + 1))
    vColTotals = []
    for i in range(4):
        vColTotals.append([0] * (g_lXTabColEndPlanHoriz + 1))
        
    vColTotals[0][0] = 'On-Hand (UU)'
    vColTotals[1][0] = 'Entradas (UU)'
    vColTotals[2][0] = 'Venta (UU)'
    vColTotals[3][0] = 'Venta Perdida (UU)'
    
    # Loop through each product and check it is in scope
    # --------------------------------------------------
    lNumDRPTitleBlocks = 0
    # lProdStoreRow = 1
    for lProdStoreRow in np.arange(1, g_lNumProdStoreRows + 1):
        if g_prodStoreRowSelected[lProdStoreRow-1] :
            # Populate Title Blocks
            lNumDRPTitleBlocks = lNumDRPTitleBlocks + 1
            lStoreRow = g_lStoreRowForProdStoreRow[lProdStoreRow - 1]
            lProdRow = g_lProdRowForProdStoreRow[lProdStoreRow - 1]

            lFirstLine = 1 + (lNumRowsPerDRPTitleBlock + 1) * (lNumDRPTitleBlocks - 1) - 1

            vDRPTitleBlocks[lFirstLine][0] = g_productDims[lProdRow - 1][0] + ' - ' + g_productDims[lProdRow - 1][1]

            vDRPTitleBlocks[lFirstLine + 2][0] = "Sucursal: " + str(g_storesDims[lStoreRow - 1][1])

            vDRPTitleBlocks[lFirstLine + 4][0] = "Stock Inicial (UU)"
            vDRPTitleBlocks[lFirstLine + 4][1] = g_lStartingInvForProdStoreRow[lProdStoreRow - 1]

            vDRPTitleBlocks[lFirstLine + 5][0] = "Lead Time (Dias)"
            vDRPTitleBlocks[lFirstLine + 5][1] = g_lLeadTimeForStoreRow[lStoreRow - 1]

            vDRPTitleBlocks[lFirstLine + 6][0] = "Cobertura (Dias)"
            vDRPTitleBlocks[lFirstLine + 6][1] = g_lDaysCoverForProdStoreRow[lProdStoreRow - 1]

            vDRPTitleBlocks[lFirstLine + 7][0] = "Pedido Min. / Mult. (UU)"
            vDRPTitleBlocks[lFirstLine + 7][1] = g_lMinLotSizeForProdStoreRow[lProdStoreRow - 1]

            vDRPTitleBlocks[lFirstLine + 8][0] = "Tamano Lote (UU)"
            vDRPTitleBlocks[lFirstLine + 8][1] = g_lLotSizeForProdStoreRow[lProdStoreRow - 1]

            vDRPTitleBlocks[lFirstLine + 9][0] = "EOQ (UU)"
            vDRPTitleBlocks[lFirstLine + 9][1] = g_lEOQForProdStoreRow[lProdStoreRow - 1]

            # Populate time-varying data
            # --------------------------

            for lXTabCol in np.arange(g_lXTabColToday, g_lXTabColEndPlanHoriz + 1):
                # Populate main blocks
                # --------------------
                vDRPBlocks[lFirstLine][0] = "Plan de Demanda"
                vDRPBlocks[lFirstLine + 1][0] = "Stock Objetivo (al fin del día)"
                vDRPBlocks[lFirstLine + 2][0] = "Stock On-Hand (al fin del día)"
                vDRPBlocks[lFirstLine + 3][0] = "En Transito (al fin del día)"
                vDRPBlocks[lFirstLine + 4][0] = "Entradas Confirmadas"
                vDRPBlocks[lFirstLine + 5][0] = "Pedidos Planeadas"
                vDRPBlocks[lFirstLine + 6][0] = "Despachos Planeados"
                vDRPBlocks[lFirstLine + 7][0] = "Entradas Planeadas"
                vDRPBlocks[lFirstLine + 8][0] = "Venta"
                vDRPBlocks[lFirstLine + 9][0] = "Venta Perdida"

                vDRPBlocks[lFirstLine][lXTabCol] = g_dPlanDemForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]  # Plan de Demanda
                vDRPBlocks[lFirstLine + 1][lXTabCol] = g_lTgtStockForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]  # Stock Objetivo
                vDRPBlocks[lFirstLine + 2][lXTabCol] = g_lPlanInvForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]  # Stock On Hand
                vDRPBlocks[lFirstLine + 3][lXTabCol] = g_lPlanInTransitForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]  # Stock en Tránsito
                vDRPBlocks[lFirstLine + 4][lXTabCol] = g_lSchedRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]  # Entradas confirmadas
                vDRPBlocks[lFirstLine + 5][lXTabCol] = g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]  # Pedidos Planeados
                vDRPBlocks[lFirstLine + 6][lXTabCol] = g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]  # Despachos Planeados
                vDRPBlocks[lFirstLine + 7][lXTabCol] = g_lPlanRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]  # Entradas Planeadas
                vDRPBlocks[lFirstLine + 8][lXTabCol] = g_dPlanDemForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] - \
                                                       g_lPlanRefusForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]  # 'Venta
                vDRPBlocks[lFirstLine + 9][lXTabCol] = g_lPlanRefusForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]  # Venta Perdida

            # Stop if max number of blocks reached
            # ------------------------------------

            if lNumDRPTitleBlocks == lMaxDRPTitleBlocks:
                print("Se muestran solo los primeros " + str(lNumDRPTitleBlocks) + " productos")
                break

    # Calculate aggregate metrics
    # ---------------------------
    # lProdStoreRow=1
    for lProdStoreRow in np.arange(1, g_lNumProdStoreRows + 1):
        if g_prodStoreRowSelected[lProdStoreRow-1]:
            # Populate Title Blocks
            lStoreRow = g_lStoreRowForProdStoreRow[lProdStoreRow - 1]
            lProdRow = g_lProdRowForProdStoreRow[lProdStoreRow - 1]

            g_productDims_index = int(g_productDims[lProdRow - 1][0])
            dFactor = g_dictConvFactorForProdUOM1UOM2[str(g_productDims_index) + g_dictUOMCDForProd[g_productDims_index] + g_dictMinUOMForProd[g_productDims_index]]
            #dFactor = g_lMinLotSizeForProdStoreRow[lProdStoreRow - 1]
            if dFactor > 0:
                dUnitCost = g_lUnitCostDefaultForProdRow[lProdRow - 1] / dFactor
            else:
                dUnitCost = 0

            lTotSchedRecUU, lTotPlanRecUU, lTotDispUU, \
            lTotDispPesos, lTotRefusUU, lTotInvUU, lTotDispLineas, \
            dTotInvPesos, dTotInTransitPesos = planeación_detallada_temp(g_lXTabColToday, g_lXTabColEndPlanHoriz,
                                                 lTotSchedRecUU, g_lSchedRecForProdStoreRowXTabCol,
                                                 lTotPlanRecUU, g_lPlanRecForProdStoreRowXTabCol, 
                                                 lProdStoreRow, lTotDispUU, g_lPlanOrdForProdStoreRowXTabCol,
                                                 dFactor, lTotDispPesos, dUnitCost, lTotRefusUU, lTotInvUU,
                                                 g_lPlanInvForProdStoreRowXTabCol, lTotDispLineas, dTotInvPesos,
                                                 dTotInTransitPesos)
#             for lXTabCol in np.arange(g_lXTabColToday, g_lXTabColEndPlanHoriz + 1):
#                 # TOTALES
#                 # -------

#                 # Entradas confirmadas
#                 lTotSchedRecUU[lXTabCol] = lTotSchedRecUU[lXTabCol] + g_lSchedRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
#                 # Entradas PlaneadaS
#                 lTotPlanRecUU[lXTabCol] = lTotPlanRecUU[lXTabCol] + g_lPlanRecForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
#                 # Pedidos Planeados O Despachos Planeados en Unidades Cedi
#                 if dFactor > 0:
#                     lTotDispUU[lXTabCol] = lTotDispUU[lXTabCol] + g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] / dFactor

#                 # Total Distribucion en $MM
#                 lTotDispPesos[lXTabCol] = lTotDispPesos[lXTabCol] + g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] * (dUnitCost / 1000000)
#                 # Venta Perdida
#                 lTotRefusUU[lXTabCol] = lTotRefusUU[lXTabCol] + g_lPlanRefusForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
#                 # Stock On Hand
#                 lTotInvUU[lXTabCol] = lTotInvUU[lXTabCol] + g_lPlanInvForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol]
#                 # Total Lineas
#                 if g_lPlanOrdForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] > 0:
#                     lTotDispLineas[lXTabCol] = lTotDispLineas[lXTabCol] + 1

#                 # Total de Inventarios en manos en $MM
#                 dTotInvPesos[lXTabCol] = dTotInvPesos[lXTabCol] + g_lPlanInvForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] * dUnitCost
#                 # Total de Inventarios en transito en $MM
#                 dTotInTransitPesos[lXTabCol] = dTotInTransitPesos[lXTabCol] + g_lPlanInTransitForProdStoreRowXTabCol[lProdStoreRow - 1][lXTabCol] * dUnitCost

        # Populate totals on DRP tab
        # --------------------------
        
    
        vColTotals[1][g_lXTabColToday:g_lXTabColEndPlanHoriz + 1] = lTotSchedRecUU[g_lXTabColToday:g_lXTabColEndPlanHoriz + 1] + lTotPlanRecUU[g_lXTabColToday:g_lXTabColEndPlanHoriz + 1]
        
    vColTotals[0][g_lXTabColToday:g_lXTabColEndPlanHoriz + 1] = lTotInvUU[g_lXTabColToday:g_lXTabColEndPlanHoriz + 1]
    vColTotals[2][g_lXTabColToday:g_lXTabColEndPlanHoriz + 1] = lTotDispUU[g_lXTabColToday:g_lXTabColEndPlanHoriz + 1]
    vColTotals[3][g_lXTabColToday:g_lXTabColEndPlanHoriz + 1] = lTotRefusUU[g_lXTabColToday:g_lXTabColEndPlanHoriz + 1]
        
#         for lXTabCol in np.arange(g_lXTabColToday, g_lXTabColEndPlanHoriz + 1):
#             vColTotals[0][0] = 'On-Hand (UU)'
#             vColTotals[1][0] = 'Entradas (UU)'
#             vColTotals[2][0] = 'Venta (UU)'
#             vColTotals[3][0] = 'Venta Perdida (UU)'

#             vColTotals[0][lXTabCol] = lTotInvUU[lXTabCol]
#             vColTotals[1][lXTabCol] = lTotSchedRecUU[lXTabCol] + lTotPlanRecUU[lXTabCol]
#             vColTotals[2][lXTabCol] = lTotDispUU[lXTabCol]
#             vColTotals[3][lXTabCol] = lTotRefusUU[lXTabCol]
        
        
    # Set up column headers
    # ---------------------
    vColHeaders = []
    for i in range(3):
        vColHeaders.append([''] * (g_lXTabColEndPlanHoriz + 1))
        
#     vColHeaders = np.empty([3, g_lXTabColEndPlanHoriz + 1], dtype="S40")

    for lXTabCol in np.arange(g_lXTabColToday, g_lXTabColEndPlanHoriz + 1):
#         vColHeaders[1][lXTabCol] = datetime.date(g_lDateForXTabCol[lXTabCol]).weekday()
        vColHeaders[1][lXTabCol] = g_lDateForXTabCol_dayofweek[lXTabCol]
#         vColHeaders[2][lXTabCol] = datetime.date(g_lDateForXTabCol[lXTabCol]).day
        vColHeaders[2][lXTabCol] = g_lDateForXTabCol_day[lXTabCol]

        if vColHeaders[1][lXTabCol] == 0 : vColHeaders[1][lXTabCol] = "Lu"
        if vColHeaders[1][lXTabCol] == 1 : vColHeaders[1][lXTabCol] = "Ma"
        if vColHeaders[1][lXTabCol] == 2 : vColHeaders[1][lXTabCol] = "Mi"
        if vColHeaders[1][lXTabCol] == 3 : vColHeaders[1][lXTabCol] = "Ju"
        if vColHeaders[1][lXTabCol] == 4 : vColHeaders[1][lXTabCol] = "Vi"
        if vColHeaders[1][lXTabCol] == 5 : vColHeaders[1][lXTabCol] = "Sa"
        if vColHeaders[1][lXTabCol] == 6 : vColHeaders[1][lXTabCol] = "Do"

        if lXTabCol == g_lXTabColToday or vColHeaders[2][lXTabCol] == 1 :
#             vColHeaders[0][lXTabCol] = datetime.date(g_lDateForXTabCol[lXTabCol]).month
            vColHeaders[0][lXTabCol] = g_lDateForXTabCol_month[lXTabCol]

            if vColHeaders[0][lXTabCol] == 1: vColHeaders[0][lXTabCol] = "Enero"
            if vColHeaders[0][lXTabCol] == 2: vColHeaders[0][lXTabCol] = "Febrero"
            if vColHeaders[0][lXTabCol] == 3: vColHeaders[0][lXTabCol] = "Marzo"
            if vColHeaders[0][lXTabCol] == 4: vColHeaders[0][lXTabCol] = "Abril"
            if vColHeaders[0][lXTabCol] == 5: vColHeaders[0][lXTabCol] = "Mayo"
            if vColHeaders[0][lXTabCol] == 6: vColHeaders[0][lXTabCol] = "Junio"
            if vColHeaders[0][lXTabCol] == 7: vColHeaders[0][lXTabCol] = "Julio"
            if vColHeaders[0][lXTabCol] == 8: vColHeaders[0][lXTabCol] = "Agosto"
            if vColHeaders[0][lXTabCol] == 9: vColHeaders[0][lXTabCol] = "Septiembre"
            if vColHeaders[0][lXTabCol] == 10: vColHeaders[0][lXTabCol] = "Octubre"
            if vColHeaders[0][lXTabCol] == 11: vColHeaders[0][lXTabCol] = "Noviembre"
            if vColHeaders[0][lXTabCol] == 12: vColHeaders[0][lXTabCol] = "Diciembre"


    # NOTA:
    # Las matrices que contienen las tablas que se muestran en la hoja "Planeacion Detallada" de la version Excel son:
    # - vDRPTitleBlocks : Es el Detalle de cada producto-tienda
    # - vDRPBlocks : Es el la información totalizada con la que se hace la gráfica
    # - vColTotals : Tiene la información diaria detallada de la plneación de cada producto tienda
    # - vColHeaders : Tiene las cabeceras con meses, dias de semana y dia

    # La conversión a Dataframes no es parte del algoritmo, solo se hace con fines didacticos.
    return vDRPTitleBlocks, vDRPBlocks, vColTotals, vColHeaders
#     df_vDRPTitleBlocks = pd.DataFrame(vDRPTitleBlocks)
#     df_vDRPBlocks = pd.DataFrame(vDRPBlocks)
#     df_vColTotals = pd.DataFrame(vColTotals)
#     df_vColHeaders = pd.DataFrame(vColHeaders)
    
vDRPTitleBlocks, vDRPBlocks, vColTotals, vColHeaders = RealityCheck(g_prodStoreRowSelected, g_lDateForXTabCol_dayofweek, 
                                                         g_lDateForXTabCol_day, g_lDateForXTabCol_month,
                                                         g_lXTabColToday, g_lXTabColEndPlanHoriz, g_lNumProdStoreRows, 
                                                         g_productDims, g_lSchedRecForProdStoreRowXTabCol, 
                                                         g_lPlanRecForProdStoreRowXTabCol, g_lPlanOrdForProdStoreRowXTabCol,
                                                         g_lPlanRefusForProdStoreRowXTabCol, g_lPlanInvForProdStoreRowXTabCol,
                                                         g_lPlanInTransitForProdStoreRowXTabCol)
```
