```
import nltk

from nltk.book import *
from nltk.corpus import stopwords
from string import punctuation


blake_poems = nltk.corpus.gutenberg.words('blake-poems.txt')
# len(blake_poems)

stop_words = list(set(stopwords.words('english')))
stop_words_2 = [w.capitalize() for w in stop_words]
stop_words_3 = [w.upper() for w in stop_words]

stop_words_all = stop_words + stop_words_2 + stop_words_3
# stop_words_all[:2]

ans = [w for w in blake_poems if w not in stop_words_all]
ans = [w for w in ans if w not in punctuation] #if needed
ans = [w for w in ans if w not in '.""'] # custom character to remove
ans = FreqDist(ans)

ans.most_common()[:25]
```
