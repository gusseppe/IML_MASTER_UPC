**Knowledge in Speech and Language Processing:**

- Morphology: variations of individuals words (e.g door, doors). Meaningful components of words.
- Syntax: order or group words together (I have cat  a. I have a cat). Structural relationship btw words.
- Lexical semantics: understand the meaning of the words.
- coreference resolution: understand words based on previous sentences (e.g that, she. Where is she?).

**Ambiguity:**

- Some imput is amiguous if there are many alternative structure than can be built for it.
- For instance, the sentence: I made her duck, has many meanings.
  - I cocked a duck for her.
  - I cooked a duck belonging to her.
  
- Deciding whether "duck" is a noun or a verb can be solved by **Part-of-Speech-tagging (POS)**. Deciding whether 'make' means 'create' or 'cook' can be solved by **word sense disambiguation**.

**Models and Algorithms**

- State Machine: consists in states, transitions among states and input representation.
- Variations fo states machines are: (non) deterministic finite-state automata, finite-state transducers.
- Markov model: State Machine + probabilities.

**Language, Thought, and Understanding**

-The ability of computers to process language like humans, is a signal of truly intelligent machines.
- If a machine can pass the Turing test (Alan Turing, 1950) then it would be considered intelligent.
- The Turing test consists of 2 people and 1 computer. One person has the role of an interrogator. The other person with the computer have the rol of contestans. So that, the interrogator ask several questions by teletype. The task of the machine is to fool the interrogator by answering rational and natural responses.The task of the second person is to convince the interrogator that the other partipant is the machine and she is a human. So, if the interrogator can diferentiate whether one of them if a computer or a human, then the computer passes the test.

### Automatic Natural Language Understanding

A major part of intelligent behaviour is understanding language to build intelligent machines.

**Word Sense Disambiguation**

The goal is finding out the sense of a word in a given context.

For instance:

a. The lost children were found by the searchers (agentive)

b. The lost children were found by the mountain (locative)

c. The lost children were found by the afternoon (temporal)

**Pronoun Resolution**

To detect "who did what to whom", like subjects and objects of verbs.

a. The thieves stole the paintings. They were subsequently sold.

b. The thieves stole the paintings. They were subsequently caught.

c. The thieves stole the paintings. They were subsequently found.

Techniques to solve that:
  - ana-phora resolution: to find what a pronoun or noun phrase refers to.
  - sematic role labeling: how a noun phrase relates to the verb.


**Generating Language Output**

Having solved the previous problems, the *question answering* can be also solved.

a. Text: ... The thieves stole the paintings. They were subsequently sold. ...

b. Human: Who or what was sold?

c. Machine: The paintings.


**Machine Translation**

### Counting Words by Genre

````
from nltk.corpus import brown

cfd = nltk.ConditionalFreqDist(
        (genre, word) 
        for genre in brown.categories()
        for word in brown.words(categories=genre)
      )

cfd
````

````
cfd.conditions()
````

````
adv = cfd['adventure']
adv
````

````
adv.most_common()[:10]
````
